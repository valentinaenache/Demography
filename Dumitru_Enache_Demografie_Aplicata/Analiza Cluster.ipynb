```python
import pandas as pd
import numpy as np
from sklearn.preprocessing import MinMaxScaler
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("demografie.csv")
df
```

```python
plt.scatter(df['exp_sp'],df['mult_jobs'],df['unex_f_exp'],df['inab_holiday'])
```

```python
km = KMeans(n_clusters=3 )
km
```

```python
y_predicted = km.fit_predict(df[['exp_sp','mult_jobs','unex_f_exp','inab_holiday']])
y_predicted
```

```python
df['cluster'] = y_predicted
df
```

```python
cluster_map = pd.DataFrame()
cluster_map['Tara'] = df.Tara
cluster_map['exp_sp'] = df.exp_sp
cluster_map['mult_jobs'] = df.mult_jobs
cluster_map['unex_f_exp'] = df.unex_f_exp
cluster_map['inab_holiday'] = df.inab_holiday
cluster_map['cluster'] = km.labels_
```
